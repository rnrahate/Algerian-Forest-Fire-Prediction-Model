# Data Cleaning

```python
# missing values
df[df.isnull().any(axis=1)]
```

### OHE of these two regions 
### 0-121 ==> 0 (BEJAJA Region)
### 122- ==> 1 (SIDI_BEL_ABBES Region)

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
# remove null values
df = df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# removing this repeated or duplicate row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# classes has unnecessary spaces
# fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# all datasets columns data-type must be either int or float
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.columns
```

```python
# changing the other columns to float datatype
objects = [features for features in df.columns if df[features].dtypes=='O']

for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

### cleaned and changed datatypes of columns to int and float

```python
# save the cleaned dataset
df.to_csv('Datasets/Algerian_forest_fires_dataset_CLEANED.csv',index=False)
```

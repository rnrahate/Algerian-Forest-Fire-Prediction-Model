```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
%matplotlib inline
```

```python
df = pd.read_csv('Datasets/Algerian_forest_fires_dataset_CLEANED.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

### Features Selection

```python
# Independent and Dependent features
X = df.drop('FWI',axis=1)
Y = df['FWI']
```

```python
X.head()
```

```python
Y.head()
```

### Model prepration by data splitting

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
# feature selection based on correlation
X_train.corr()
```

```python
# check for multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation (dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85
X_train.drop(corr_features, axis=1, inplace=True)
X_test.drop(corr_features, axis=1, inplace=True)
X_train.shape,X_test.shape
```

### Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

### Boxplot to understand the effect of StandardScaler()

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

## Model Training
### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train_scaled,Y_train)
Y_pred = linreg.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
mae = mean_absolute_error(Y_test,Y_pred)
score = r2_score(Y_test,Y_pred)
print("Mean Absolute Error: ",mae)
print("r2_score: ",score)
plt.scatter(Y_test,Y_pred)
# add best-fit line
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

#### above we saw that our Y_test and Y_pred, i.e. actual value and prredicted value by model has linear relation
#### which may also point to a possibility of overfitting of model. So,
### Lasso Regression Algorithm

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,Y_train)
y_pred=lasso.predict(X_test_scaled)
mae = mean_absolute_error(Y_test,y_pred)
score = r2_score (Y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter (Y_test,y_pred)
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

### Hyperparameter Tuning
#### Lasso Cross Validation

```python
from sklearn.linear_model import LassoCV
lassoCV = LassoCV(cv=5)
lassoCV.fit(X_train_scaled,Y_train)
```

```python
y_pred = lassoCV.predict(X_test_scaled)
mae = mean_absolute_error(Y_test,y_pred)
score = r2_score (Y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter (Y_test,y_pred)
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

### Ridge Regression Algorithm

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge = Ridge()
ridge.fit(X_train_scaled,Y_train)
y_pred=ridge.predict(X_test_scaled)
mae = mean_absolute_error(Y_test,y_pred)
score = r2_score (Y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter (Y_test,y_pred)
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

### Hyperparameter Tuning
#### Ridge Cross Validation

```python
from sklearn.linear_model import RidgeCV
ridgeCV = RidgeCV(cv=5)
ridgeCV.fit(X_train_scaled,Y_train)
```

```python
y_pred = ridgeCV.predict(X_test_scaled)
mae = mean_absolute_error(Y_test,y_pred)
score = r2_score (Y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter (Y_test,y_pred)
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

### ElasticNet Regression Algorithm

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic = ElasticNet()
elastic.fit(X_train_scaled,Y_train)
y_pred=elastic.predict(X_test_scaled)
mae = mean_absolute_error(Y_test,y_pred)
score = r2_score (Y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter (Y_test,y_pred)
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

### Hyperparameter Tuning
#### ElasticNet Cross Validation

```python
from sklearn.linear_model import ElasticNetCV
elasticCV = ElasticNetCV(cv=5)
elasticCV.fit(X_train_scaled,Y_train)
```

```python
y_pred = elasticCV.predict(X_test_scaled)
mae = mean_absolute_error(Y_test,y_pred)
score = r2_score (Y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter (Y_test,y_pred)
m, b = np.polyfit(Y_test, Y_pred, 1)
plt.plot(Y_test, m*Y_test + b, color='red')
```

### Why RidgeCV is the Best Fit?

Handles multicollinearity effectively using L2 regularization.

Prevents overfitting while keeping all features.

Automatically selects the optimal alpha via cross-validation.

Produces stable coefficients and consistently better predictive performance compared to Lasso and ElasticNet.
